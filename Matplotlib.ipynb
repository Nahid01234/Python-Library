### 2.Barchart

```python
id_no1=[116,160,162,163]
id_no2=[68,72,73,78]
info=["S","P","N","D"]
xp=np.arange(len(info))
plt.xticks(xp,info)
#plt.bar(xp+.2,id_no1,width=0.45,label="id_no1")
plt.bar(xp-.2,id_no2,width=0.45,label="id_no2")
plt.legend()
plt.show()
```

```python
id_no1=[116,160,162,163]
id_no2=[68,72,73,78]
info=["S","P","N","D"]
xp=np.arange(len(info))
plt.xticks(xp,info)
plt.bar(xp+.2,id_no1,color="b",width=0.45)
plt.bar(xp-.2,id_no2,width=0.45,color="g")
plt.legend(["id_no1","id_no2"])
plt.show()
```

### 5. 3D Scatter Plots

```python
fig=plt.figure()
ax=plt.subplot(projection="3d")
ax.scatter3D(data["BloodPressure"],data["Age"],data["BMI"])
ax.set_xlabel("BloodPressure")
ax.set_ylabel("Age")
ax.set_zlabel("BMI")
plt.show()
```

### 6.3D Line Plots

```python
x=[1,2,3,4]
y=[5,6,7,8]
z=[10,111,12,11]
ax=plt.subplot(projection="3d")
ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color="red")
```

### 7.3D Surface Plots

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
a,b=np.meshgrid(x,y)
c=a**2+ b**2
fig=plt.figure(figsize=(12,7))
ax=plt.subplot(projection="3d")
v=ax.plot_surface(a,b,c,cmap="viridis")
fig.colorbar(v)
```

### 8.Contour Plots

```python
fig=plt.figure(figsize=(12,7))
ax=plt.subplot()
v=ax.contour(a,b,c,cmap="viridis")
fig.colorbar(v)
```

```python
fig=plt.figure(figsize=(12,7))
ax=plt.subplot()
v=ax.contourf(a,b,c,cmap="viridis")
fig.colorbar(v)
```
